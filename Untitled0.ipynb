```python
#import all the packages

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.imdb.com/search/title/?release_date=2021&sort=num_votes,desc&page='
req = requests.get(url)
print(req)
soup = BeautifulSoup(req.text,'html')
#print(soup.prettify())
```

```python
m_tag = soup.findAll('h3',attrs={'class':'lister-item-header'})
m_tag
```

```python
movie_list = [i.a.text.strip() for i in m_tag]
movie_list
```

```python
len(movie_list)
```

```python
m_s = [i.text.strip() for i in soup.findAll('span',attrs={'class':'metascore'})]
```

```python
m_s
```

```python
len(m_s)
```

```python
pd.DataFrame({'Movie Name':movie_list,
              'MetaScore' : m_s})
```

```python
m_sc_d = soup.findAll('div',attrs={'class':'ratings-bar'})
m_sc_d
```

```python
m_sc_d[0]
```

```python
m_sc_d[0].find('span',attrs={'class':'metascore'})
```

```python
m_sc_d[1]
```

```python
print(m_sc_d[1].find('span',attrs={'class':'metascore'}))
```

```python
meta_score = []
for d in m_sc_d:
  a = d.find('span',attrs={'class':'metascore'}) 
  if a is not None:
    meta_score.append(a.text.strip())
  else:
    meta_score.append('None')
```

```python
len(meta_score)
```

```python
meta_score
```

```python
pd.DataFrame({'Movie Name':movie_list,
              'MetaScore' : meta_score})
```
